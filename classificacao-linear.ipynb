```python
import numpy as np
import matplotlib.pyplot as plt

#Equação do perceptron segue a mesma equação da reta
a = -1
b = 4
c = 0.4

#Solução da equação da reta
# ax + by + c = 0
# y = (-a*x -c)/b

x = np.linspace(-2, 4, 50)
y = (-a*x -c)/b

#Plot da reta no plano n2
plt.axvline(0, -1, 1, color='k', linewidth=1)
plt.axhline(0, -2, 4, color='k', linewidth=1)
plt.plot(x, y)
plt.grid(True)

#Plot de ponto no plano que passa pela reta
#resolução igualdade matemática n=0
p1 = (2, 0.4)
ret1 = a*p1[0] + b*p1[1] + c
plt.plot(p1[0], p1[1], color='b', marker='o')
print("blue point: %.2f" % ret1)

#Plot de ponto no plano acima da reta
#resolução de diferença onde n>0
p2 = (1, 0.6)
ret2 = a*p2[0] + b*p2[1] + c
plt.plot(p2[0], p2[1], color='g', marker='o')
print("green point: %.2f" % ret2)

#Plot de ponto no plano abaixo da reta
#resolução de diferença onde n<0
p3 = (3, -0.4)
ret3 = a*p3[0] + b*p3[1] + c
plt.plot(p3[0], p3[1], color='r', marker='o')
print("red point: %.2f" % ret3)
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(46)

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)

plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolors='k')

p = X[10]
print(Y[10])
plt.plot(p[0], p[1], marker='^', markersize=15)
```

```python
def plotmodel(w1, w2, b):
    plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolors='k')

    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()
    x = np.linspace(-2, 4, 50)
    y = (-w1*x -b)/w2

    plt.axvline(0, -1, 1, color='k', linewidth=1)
    plt.axhline(0, -2, 4, color='k', linewidth=1)
    plt.plot(x, y)
    plt.grid(True)

    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)

def classify(ponto, w1, w2, b):
    ret = w1 * ponto[0] + w2 * ponto[1] + b

    if ret >= 0:
        return 0, 'blue'
    else:
        return 1, 'yellow'
    
p = (1, -1)
w1 = -3
w2 = -0.42
b = 0.5
classe, cor = classify(p, w1, w2, b)
print(classe, cor)

plotmodel(w1, w2, b)
plt.plot(p[0], p[1], marker='^', color=cor, markersize=10)

acertos = 0
for k in range(len(X)):
    categ, _ = classify(X[k], w1, w2, b)
    if categ == Y[k]:
        acertos += 1

print("Acurácia: {0}".format(100*acertos/len(X)))
```
